2. Feature Engineering

```python
# Define category columns
essential_columns = ['rent', 'groceries', 'transport', 'utilities', 'healthcare', 'education']
discretionary_columns = ['eating_out', 'entertainment', 'Miscellaneous']
```

```python
# Check if all these columns exist to avoid KeyError
all_required_columns = essential_columns + discretionary_columns
```

```python
# Assuming Total_Expenses is the sum of major expense categories
expense_cols = ['Rent', 'Loan_Repayment', 'Insurance', 'Groceries', 'Transport', 
                'Utilities', 'Healthcare', 'Eating_Out', 'Entertainment', 'Education', 'Miscellaneous']
```

```python
df_encoded['Total_Expenses'] = df_encoded[expense_cols].sum(axis=1)
```

```python
# Behavioral Ratios
df_encoded['Essentials_Ratio'] = df_encoded[['Rent', 'Groceries', 'Transport', 'Utilities', 'Healthcare']].sum(axis=1) / df_encoded['Total_Expenses']
df_encoded['Discretionary_Ratio'] = df_encoded[['Entertainment', 'Eating_Out', 'Miscellaneous']].sum(axis=1) / df_encoded['Total_Expenses']
df_encoded['Loan_Stress_Ratio'] = df_encoded['Loan_Repayment'] / df_encoded['Total_Expenses']
df_encoded['Insurance_Coverage_Ratio'] = df_encoded['Insurance'] / df_encoded['Total_Expenses']
df_encoded['Healthcare_Weight'] = df_encoded['Healthcare'] / df_encoded['Total_Expenses']
```

```python
# Financial Health Score
df_encoded['Financial_Health_Score'] = (
    0.3 * (1 - df_encoded['Loan_Stress_Ratio']) +
    0.3 * df_encoded['Essentials_Ratio'] +
    0.2 * (1 - df_encoded['Discretionary_Ratio']) +
    0.2 * df_encoded['Insurance_Coverage_Ratio']
)
```

```python
recommendations = {
    0: "You’re quite frugal. Consider exploring investment opportunities or higher-interest savings accounts.",
    1: "You tend to overspend. Try setting monthly budgets and using expense trackers like Mint or YNAB.",
    2: "You have a balanced profile. Maintain your habits and explore diversified investment options.",
    3: "Your spending is inconsistent. Consider setting clear financial goals and monitoring your cash flow.",
}
```

Predict and Recommend

```python
def get_behavior_recommendation(user_input, model, encoder, recommendation_dict):
    input_df = pd.DataFrame([user_input])

    # Get model's expected features
    model_features = model.feature_names_in_

    # Check for missing features
    missing = [col for col in model_features if col not in input_df.columns]
    if missing:
        raise ValueError(f"Missing features in user input: {missing}")

    # Filter and reorder columns to match model
    input_df = input_df[model_features]

    prediction = model.predict(input_df)[0]
    advice = recommendation_dict.get(prediction, "No recommendation found for this type.")
    return prediction, advice
```

6. Expense Forecasting

```python
# forecaster.py

from prophet import Prophet
import pandas as pd

def forecast_next_month(data: list[dict], periods: int = 1):
    """
    Forecasts next month's expense from historical category-wise data.
    Input: list of dicts with 'date' and 'amount'
    Output: Forecasted value for next month
    """
    # Convert to DataFrame
    df = pd.DataFrame(data)
    df['ds'] = pd.to_datetime(df['date'])
    df['y'] = df['amount']
    df = df[['ds', 'y']]

    # Initialize and train model
    model = Prophet()
    model.fit(df)

    # Make future predictions
    future = model.make_future_dataframe(periods=periods, freq='MS')
    forecast = model.predict(future)

    # Extract forecast for next month
    next_month_pred = forecast[['ds', 'yhat']].tail(periods)
    return {
        "forecast": round(next_month_pred['yhat'].iloc[0], 2),
        "date": str(next_month_pred['ds'].iloc[0].date())
    }
```

```python
def generate_suggestions(income: float, current: dict, forecast: dict):
    suggestions = []

    # Rule-based checks
    if current.get("Eating_Out", 0) > 0.25 * income:
        suggestions.append("Your dining expenses exceed 25% of income. Consider reducing them.")

    if current.get("Rent", 0) > 0.5 * income:
        suggestions.append("Rent exceeds 50% of income. Explore more affordable housing.")

    if current.get("Savings", 0) < 0.1 * income:
        suggestions.append("Your savings are below 10% of income. Try to save more.")

    # Forecast comparison
    for category in current:
        if forecast.get(category, 0) > current[category] * 1.15:
            suggestions.append(f"Forecasted {category} is increasing significantly. Monitor and adjust.")
        elif forecast.get(category, 0) < current[category] * 0.85:
            suggestions.append(f"{category} spending is expected to drop. Great job managing it!")

    return suggestions
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score

def evaluate_model(y_true, y_pred, y_proba=None, average='weighted'):
    metrics = {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, average=average),
        "recall": recall_score(y_true, y_pred, average=average),
        "f1_score": f1_score(y_true, y_pred, average=average),
        "confusion_matrix": confusion_matrix(y_true, y_pred).tolist()
    }

    if y_proba is not None:
        try:
            metrics["roc_auc"] = roc_auc_score(y_true, y_proba, multi_class='ovr')
        except:
            metrics["roc_auc"] = "N/A"

    return metrics
```
